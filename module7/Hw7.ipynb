```python
import numpy as np
from surprise import Dataset, Reader
from surprise.model_selection import cross_validate
from surprise.prediction_algorithms import SVD, SVDpp, NMF

data = Dataset.load_builtin('ml-100k')

reader = Reader(rating_scale=(1, 5))
```

```python
svd = SVD()
svdpp = SVDpp()
nmf = NMF()

models = [svd, svdpp, nmf]
for model in models:
    print(f"Cross-validating {model.__class__.__name__}...")
    results = cross_validate(model, data, measures=['RMSE'], cv=5, verbose=True)
    print(f"Mean RMSE for {model.__class__.__name__}: {np.mean(results['test_rmse']):.4f}")
```
